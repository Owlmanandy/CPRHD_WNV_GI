```python
import pandas as pd
import datetime
```

```python
se = pd.read_csv("socio_economic_all_counties_1995-2009.csv")
print(se.info())
print("\nUnique Years: ", se.YEAR.nunique(), "\n", se.YEAR.unique())
print("\nUnique Counties: ", se.STCOU.nunique(), "\n")
se.head(5)
```

```python
for geoid in se.STCOU.unique():
    temp = pd.DataFrame()
    temp = se[se['STCOU'] == geoid]
    name = 'SocioEconomic_county_files/'+temp.Areaname.iloc[0]+'.csv'
    print("Processing - ", temp.Areaname.iloc[0], " dataset")
    temp.to_csv(name, index=False)
    
```

```python
import os
import glob
import pandas as pd
```

```python
path = '/Users/vpb/Documents/Projects/WNV_Challenge/SocioEconomic_county_files/Outputs/'
extension = 'csv'
os.chdir(path)
result = glob.glob('*.{}'.format(extension))
print(len(result))
```

```python
final = pd.DataFrame()
for county in result:
    temp_df = pd.read_csv(path+county)
    final = final.append(temp_df, ignore_index=True)
    
final
```

```python
final = final.sort_values(by=['STCOU', 'YEAR'])
final = final.reset_index(drop=True)
final
```
